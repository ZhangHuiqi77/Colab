```python
# plot in different colors

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plotpca3d(X,grp,titletxt):
    # add label as the last column
    # X_new is a 2D array with 3 principal components and 1 label column
    X_new = np.concatenate((X, np.array([grp]).T), axis=1)

    # separate the data by label
    X_class0 = X_new[X_new[:, -1] == 1, :-1]
    X_class1 = X_new[X_new[:, -1] == 2, :-1]
    X_class2 = X_new[X_new[:, -1] == 3, :-1]

    # create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # scatter plot the transformed data in 3D space
    ax.scatter(X_class0[:, 0], X_class0[:, 1], X_class0[:, 2], c='b', label='Class 0')
    ax.scatter(X_class1[:, 0], X_class1[:, 1], X_class1[:, 2], c='r', label='Class 1')
    ax.scatter(X_class2[:, 0], X_class2[:, 1], X_class2[:, 2], c='g', label='Class 2')


    # set the axis labels and legend
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    ax.legend()

    # show the plot
    plt.title(titletxt)
    plt.show()
```
