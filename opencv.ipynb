```python
from google.colab.patches import cv2_imshow
```

```python
capture = cv.VideoCapture('/content/gdrive/MyDrive/deeplabcut/group5_18817_bang_019.avi')

fps = capture.get(cv.CAP_PROP_FPS)
# finding the frame at a paticular time
minutes = 2
seconds = 30
frame_id = int(fps*(minutes*60 + seconds))
print('frame id =',frame_id)

# Extracting the frame from the video
capture.set(cv.CAP_PROP_POS_FRAMES, frame_id)
ret, frame = capture.read()

# Displaying and saving the frame
cv2_imshow(frame)
cv.waitKey(0)
```

```python
# play the video

while True:
    isTrue, frame = capture.read()
    cv2_imshow(frame)
    
    if cv.waitKey(20) & 0xFF==ord('d'):
        break
    
```
