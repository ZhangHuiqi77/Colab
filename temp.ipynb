```python
import numpy as np
import tensorflow as tf

# Generate some random time series data
X = np.random.rand(100, 1)

# Define the input and output sequences
X_train = X[:-1]
y_train = X[1:]

# Define the LSTM model architecture
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(64, input_shape=(1, 1)),
    tf.keras.layers.Dense(1)
])

# Compile the model with a mean squared error loss function
model.compile(loss='mse', optimizer='adam')

# Train the model on the input and output sequences
model.fit(X_train.reshape(-1, 1, 1), y_train, epochs=10)

# Use the model to predict the next value in the time series
next_value = model.predict(np.array([[X[-1]]]))
print(next_value)
```
